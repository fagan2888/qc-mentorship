Mott insulator gap on single site model with 2 sliders for $U$ and $T$. 

$-1 \leq U \leq 4$ for $T = .5, 2$

```python
import numpy as np 

def density(T, U, mu): 
    B = 1/T
    Z = 1 + 2*np.exp(B * mu) + np.exp(-B*U + 2*B*mu)
    return 1/Z * 2 * (np.exp(B * mu) + np.exp(2 * B * mu - B * U))
```

```python
import plotly.graph_objects as go 

# Ranges 
U_range = np.arange(-1, 8, 0.1)
x_range = np.arange(-2, 10, 0.1)

# Create figure 
fig = go.Figure()
fig.update_xaxes(tick0=0, dtick=1)
# fig.update_yaxes(tick0=0, dtick=1/6, range=[0, 2])
fig.update_yaxes(range=[0,2])

# Add axes 
fig.update_layout(
#     title={
#         "text": "Mott Insulating Gap on Single Site Hubbard Model", 
#         "xanchor": "center", 
#         "yanchor": "top"
#     }, 
    title="Mott Insulating Gap on Single Site Hubbard Model",
    xaxis_title=r"Chemical potential <i>μ</i>", 
    yaxis_title=r"Density <i>ρ</i>", 
)

# Add a trace for each slider step 
for step_U in U_range:
    fig.add_trace(
        go.Scatter(
            visible=False, 
            line=dict(color="#000000", width=2),
            hoverinfo='x+y',
            name='T = 0.5 Kelvin',
            # The aranges have to match!
            x=x_range, 
            y=density(0.5, step_U, x_range)
        )
    )

for step_U in U_range:
    fig.add_trace(
        go.Scatter(
            visible=False, 
            line=dict(color="#FF0000", width=2),
            hoverinfo='x+y',
            name='T = 2 Kelvin',
            # The aranges have to match!
            x=x_range, 
            y=density(2, step_U, x_range)
        )
    )
    

# Make 10th trace visible 
fig.data[10].visible = True 
fig.data[10 + len(U_range)].visible = True


# Precompute data  
steps = []
for i in range(len(U_range)):
    step = dict(
        method="update",
        args=[{"visible": [False] * len(fig.data)}], 
        label=round(U_range[i], 2)
    )
    step['args'][0]['visible'][i] = True
    step['args'][0]['visible'][i + len(U_range)] = True
    steps.append(step)

# Create sliders 
sliders = [
    dict(
        active=10, 
        currentvalue={"prefix": "Potential energy <i>U</i> (Joules): "}, 
        pad={"t": 50}, 
        steps=steps
    )
]

fig.update_layout(sliders=sliders)


# fig.write_html('mott.html', auto_open=True, include_plotlyjs='cdn')
fig.write_html('mott.html', auto_open=False, full_html=False, include_plotlyjs='cdn')
# fig.show()
```
